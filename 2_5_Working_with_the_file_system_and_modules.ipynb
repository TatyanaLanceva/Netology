Переведите содержимое файла purchase_log.txt* в словарь purchases вида:
{‘1840e0b9d4’: ‘Продукты’, …}

Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки, если покупка была, сам файл visit_log.csv* изменять не надо. Запишите в файл funnel.csv визиты из файла visit_log.csv*, в которых были покупки с указанием категории.
Учтите условия на данные:
*   содержимое purchase_log.txt* помещается в оперативную память компьютера;
*   содержимое visit_log.csv* — нет; используйте только построчную обработку этого файла.


## Задание 1.

```python
# Переводим содержимое файла purchase_log.txt* в словарь purchases по заданному шаблону.

import json
purchase= open('purchase_log.txt').readlines()

# Создаем словарь
purchases = {}

# Задаем user_id и category, делаем очистку

for line in purchase:
  line = line.strip()
  dict_=json.loads(line)
  user_id = dict_['user_id']
  category = dict_['category']
  purchases.setdefault(user_id,category)

# Выводим тип
print(type(purchases))
```

```python
# Проверим данные, выведем 5 сторк.

i = 0
for user_id,category in purchases.items():
  line = line.strip()
  print(f'{user_id}:{category}')
  i +=1
  if i >5:
    break
```

# Задание 2.

```python
# Используем модуль csv, чтобы упростить работу с разделителями и символами

import csv

# Открываем файл visit_log.csv (v) для чтения, а funnel.csv (f) - для записи, указываем кодировку

with open('visit_log.csv', newline='', encoding='utf-8') as v, \
     open('funnel.csv', 'w', newline='', encoding='utf-8') as f:

    reader = csv.reader(v)
    writer = csv.writer(f)

# Сопоставляем user_id со словарем purchases, записываем третий столбец, чистим лишние символы:

    for row in reader:
        user_id = row[0].strip()
        source = row[1].strip()

        if user_id in purchases:
            category = purchases.get(user_id, '')
            writer.writerow([user_id, source, category])
```

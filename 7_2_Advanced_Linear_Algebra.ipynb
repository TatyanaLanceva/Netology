````
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````
### YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt

# Исходный объект
A = np.array([
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
])

# 1. Матрица масштабирования (уменьшение в 2 раза)
scale_matrix = np.array([
    [0.5, 0, 0],
    [0, 0.5, 0],
    [0, 0, 1]
])
A_scaled = A @ scale_matrix

# 2. Матрица поворота (130 градусов)
theta = np.radians(130)
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta), 0],
    [np.sin(theta), np.cos(theta), 0],
    [0, 0, 1]
])
A_rotated = A @ rotation_matrix

# 3. Матрица отражения (относительно y = x)
reflect_matrix = np.array([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 1]
])
A_reflected = A @ reflect_matrix  # Применение отражения

# Извлечение координат для всех преобразований
x_orig, y_orig = A[:, 0], A[:, 1]
x_scaled, y_scaled = A_scaled[:, 0], A_scaled[:, 1]
x_rotated, y_rotated = A_rotated[:, 0], A_rotated[:, 1]
x_reflected, y_reflected = A_reflected[:, 0], A_reflected[:, 1]

# Визуализация всех преобразований
plt.figure(figsize=(10, 10))
plt.plot(x_orig, y_orig, label='Оригинал', color='blue')
plt.plot(x_scaled, y_scaled, label='Уменьшенный вполовину', color='orange')
plt.plot(x_rotated, y_rotated, label='Повернут на 130°', color='red')
plt.plot(x_reflected, y_reflected, label='Отражен относительно y=x', color='green')

# Настройки графики
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.gca().set_aspect('equal')
plt.title('Преобразования')
plt.show()

````

Найдите спектральное разложение матрицы:
```
m = np.array([[1, 2],
              [2, 3]])
print(m)
```



````
### YOUR CODE HERE

# Исходная матрица
M = np.array([[1, 2], [2, 3]])

# Собственные значения и векторы
eigenvalues, eigenvectors = np.linalg.eigh(M)

# Построение матриц P и D
D = np.diag(eigenvalues)  # Диагональная матрица из собственных значений
P = eigenvectors          # Матрица из собственных векторов

# Проверка разложения: M = P D P^T
reconstructed_M = P @ D @ P.T

# Проверка ортогональности матрицы P: P^T P = I
orthogonality_check = P.T @ P

# Округление для удобства сравнения
reconstructed_M_rounded = np.round(reconstructed_M, decimals=5)
M_rounded = np.round(M, decimals=5)

# Вывод результатов
print('Исходная матрица M:\n', M_rounded)
print('\nВосстановленная матрица P D P^T:\n', reconstructed_M_rounded)
print('\nСобственные значения:\n', eigenvalues)
print('\nМатрица собственных векторов P:\n', P)
print('\nПроверка ортогональности P^T P:\n', orthogonality_check)
````

Вам необходимо написать функцию, которая будет основана на поиске по сайту habr.com.
Функция в качестве параметра должна принимать список запросов для поиска (например, ['python', 'анализ данных']) и на основе материалов, попавших в результаты поиска по каждому запросу, возвращать датафрейм вида:

<дата> - <заголовок> - <ссылка на материал>

В рамках задания предполагается работа только с одной (первой) страницей результатов поисковой выдачи для каждого запроса. Материалы в датафрейме не должны дублироваться, если они попадали в результаты поиска для нескольких запросов из списка.


```python
import pandas as pd  # Импорт библиотеки Pandas для работы с данными
import requests  # Импорт библиотеки Requests для отправки HTTP-запросов
from bs4 import BeautifulSoup  # Импорт библиотеки BeautifulSoup для парсинга HTML
import time # Импорт библиотеки Time для работы с форматом времени
```

```python
def get_links(query):
    habr = pd.DataFrame()               # Инициализация пустого датафрейма для заполнения результатами
    URL = 'https://habr.com/ru/search/' # Указываем ссылку на поиск Хабра
    for q in query:                     # Создаем параметры для get-запроса
        params = {
            'q': q
        }
        req = requests.get(URL, params=params)                                  # Отправляем GET-запрос и получаем HTML-кода страницы
        soup = BeautifulSoup(req.text)                                          # Парсинг HTML-кода с помощью BeautifulSoup
        articles = soup.find_all('div', class_='tm-article-snippet')            # Ищем все блоки статей на странице
        for article in articles:
            title = article.find('h2').text                                     # Извлекаем заголовки
            link = article.find('a', 'tm-title__link').get('href')              # Извлекаем ссылки на статью
            date = pd.to_datetime(article.find('time').get('datetime'))         # Извлекаем даты публикации и приводим ее к типу datetime
            row = {'дата': date, 'заголовок': title, 'ссылка на материал': link}# Формирование строки с данными для добавления в датафрейм
            habr = pd.concat([habr, pd.DataFrame([row])])                       # Добавление новой строки в датафрейм
    return habr.reset_index(drop=True)                                          # Возвращаем датафрейма с результатами, сбрасываем индекс

res_1 = get_links(['python', 'анализ данных'])    # Получаем результат поиска согласно заданию по двум запросам
res_2 = res_1.drop_duplicates()                   # Очищаем результат от дубликатов
res_1                                             # Выводим результат
```

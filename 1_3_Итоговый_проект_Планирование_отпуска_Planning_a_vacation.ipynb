Загрузим данные по нашим отпускам из таблицы и выводим заголовки полученной таблицы

```python
df_journey = pd.read_csv("journey.csv", sep=",")


df_journey.head()
```

Смотрим статистику, что нет пропусков и отсутствующих значений в загруженной таблице отпусков

Добавляем в нашу загруженную таблицу отпусков данные из справочника городов. Получаем итоговую таблицу.

```python
df_merged = df_journey.merge(df_city, left_on='City', right_on='name', how='inner')
df_merged.info()
```

```python
df_merged.head()
```

Получаем датафрейм с названием города и его населением

```python
city_population = df_merged[['Город', 'population']]
city_population
```

Получим страны и среднюю численность населения в городах, в которых был отдых в последние три года

```python
country_population = df_merged[['Страна', 'Город', 'population']]
```
